```python
def process_response(response):    
    processed_response = []
    for item in response:
        dem = item[0]["generated_text"]
        dem = dem.replace("* ", "")
        dem = dem.split("Note: ")[0]
        dem = dem.split("\n\n")[0]
        processed_response.append ( dem)
    return processed_response    
        
```

### zero-shot

```python
def process_response_zero_shot(response):
    processed_response = []
    for demonstrations_one_sentence in response:
        rows_list = demonstrations_one_sentence.split("\n")

        processed_demonstrations_one_sentence = ""
        for row in rows_list:
            entity_one_row = row.split(": ")[0]
            if len(row.split(": ")) > 1:
                extracted_words_list = row.split(": ")[1].split(", ")    
            else:
                extracted_words_list = row.split(": ")[0].split(", ")    
            processed_demonstrations_one_row = ""
            for extracted_word in extracted_words_list:
                processed_demonstrations_one_row = processed_demonstrations_one_row + extracted_word + ": " + entity_one_row + '\n'
            processed_demonstrations_one_sentence = processed_demonstrations_one_sentence + processed_demonstrations_one_row 

        processed_response.append(processed_demonstrations_one_sentence)    
    
    return processed_response   
```

```python
with open("formal_eva_results_vicuna13b/thf_not_stratified_input_30_example_0_random.txt", "rb") as data:   #Pickling
    response = pickle.load(data) 
response = process_response(response)    
response = process_response_zero_shot(response)
print(response[28])    
transformed_output = change_output_format_to_tokens_tags(sentences_input, response)
transformed_solution = change_output_format_to_tokens_tags(sentences_input, demonstrations_input_solutions)
get_evaluation_without_o(transformed_solution = transformed_solution, transformed_output = transformed_output) 
```
